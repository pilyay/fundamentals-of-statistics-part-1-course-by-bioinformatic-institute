### 1.5.10 Меры изменчивости

Рассчитайте среднеквадратическое отклонение данных выборочных значений:

1 5 2 7 1 9 3 8 5 9

Не забудьте, что при расчете дисперсии и среднеквадратичного отклонения мы вычитаем единицу из общего числа наблюдений!

> [Дисперсия](https://wikimedia.org/api/rest_v1/media/math/render/svg/4bd571760d95ebeb026853456d75ffe84e605f28) (variance) - средний квадрат отклонений индивидуальных значений признака от их средний величины.

```python
_tpl = 6, 1, 2, 15, 24, 27, 30, 42, 48
tpl = 2, 4, 5, 8, 9, 10, 14, 16
_tpl = 1, 5, 2, 7, 1, 9, 3, 8, 5, 9      # объявим кортеж с заданными значениями
mean = sum(tpl) / len(tpl)              # вычислим среднее значение
print('Выборочные значения:', *tpl)
print('Всего значений:', len(tpl))
print('Сумма всех значений:', sum(tpl))
print('Среднее значение:', mean)
storage = 0
for i in range(len(tpl)):               # ...
    storage += (tpl[i] - mean) ** 2
variance = (storage / len(tpl)) ** 0.5
print('Среднеквадратическое отклонение:', variance)
```
